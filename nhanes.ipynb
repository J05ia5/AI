```python
import pandas as pd

demo = pd.read_sas("dataset_p1/3nhanes/DEMO_L.XPT")
bmx = pd.read_sas("dataset_p1/3nhanes/BMX_L.XPT")
glu = pd.read_sas("dataset_p1/3nhanes/GLU_L.XPT")
ghb = pd.read_sas("dataset_p1/3nhanes/GHB_L.XPT")
diq = pd.read_sas("dataset_p1/3nhanes/DIQ_L.XPT")
```

```python
demo = demo[["SEQN", "RIDAGEYR", "RIAGENDR"]]
bmx = bmx[["SEQN", "BMXBMI"]]
glu = glu[["SEQN", "LBXGLU"]]
ghb = ghb[["SEQN", "LBXGH"]]
diq = diq[["SEQN", "DIQ010"]]  # diabetes
```

```python
df = demo.merge(bmx, on="SEQN") \
         .merge(glu, on="SEQN") \
         .merge(ghb, on="SEQN") \
         .merge(diq, on="SEQN")
```

```python
df = df.dropna()
# 1 = sí, 2 = no → convertimos a binario
df["diabetes"] = df["DIQ010"].apply(lambda x: 1 if x == 1 else 0)

df = df.drop(columns=["DIQ010"])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = ["RIDAGEYR", "BMXBMI", "LBXGLU", "LBXGH"]

df[features] = scaler.fit_transform(df[features])
```
