```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ======================
# 1. LOAD DATA
# ======================
df = pd.read_csv("dataset_p1/4bikeSharing/train.csv")

print("Shape:", df.shape)
print(df.head())

# ======================
# 2. DATETIME FEATURES (CLAVE)
# ======================
df["datetime"] = pd.to_datetime(df["datetime"])

df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df["hour"] = df["datetime"].dt.hour
df["dayofweek"] = df["datetime"].dt.dayofweek

# Opcional (muy potente)
df["is_weekend"] = df["dayofweek"].isin([5,6]).astype(int)

# ======================
# 3. DROP COLUMN ORIGINAL
# ======================
df = df.drop(columns=["datetime"])

# ======================
# 4. MISSING VALUES
# ======================
print("\nMissing values:\n", df.isnull().sum())

# (este dataset casi no tiene NaN)

# ======================
# 5. OUTLIERS (TARGET)
# ======================
Q1 = df["count"].quantile(0.25)
Q3 = df["count"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["count"] >= Q1 - 1.5*IQR) & 
        (df["count"] <= Q3 + 1.5*IQR)]

# ======================
# 6. FEATURE ENGINEERING AVANZADO
# ======================

# Interacciones útiles
df["temp_humidity"] = df["temp"] * df["humidity"]

# Transformación log (muy importante)
df["count_log"] = np.log1p(df["count"])

# ======================
# 7. CORRELACIONES ÚTILES
# ======================
corr = df.corr(numeric_only=True)["count"].sort_values(ascending=False)

print("\nTop correlations:\n", corr.head(10))

# ======================
# 8. FEATURES / TARGET
# ======================
X = df.drop(columns=["count", "count_log"])
y = df["count_log"]  # mejor para modelos

print("Final shape:", X.shape)
```
